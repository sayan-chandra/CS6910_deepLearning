```
import matplotlib.pyplot as plt
from keras.datasets import fashion_mnist
import numpy as np
import wandb
wandb.init(project="cs6910_dl_assignment_1")
```

```
lable_names=[":t-shirt/top", ':trouser/pants', ':pullover shirt', ':dress', ':coat', ':sandal', ':shirt', ':sneaker', ':bag', ':ankle boot']
```

```

# load dataset
(trainX, trainy), (testX, testy) = fashion_mnist.load_data()
# plot first few images
b,c=plt.subplots(3,4,figsize=(9, 9))
idx=0
keepIMG=[]; captions=[]
for i in range(3):
  for j in range(4):
    val=4*i+j
    while(trainy[idx]!=val): idx+=1
    keepIMG.append(trainX[idx]); captions.append(val)
    c[i,j].imshow(trainX[idx], cmap=plt.get_cmap('gray'))
    plt.plot()
    if (i==2 and j==1): break
wandb.log(({"10 distinct fashion_mnist elements":[wandb.Image(img, caption=str(v)+kind) for img, v, kind in zip(keepIMG, captions, lable_names)]}))
wandb.finish()
```
